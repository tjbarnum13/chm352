```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define freqency points on log scale for better plotting
v=np.logspace(11,16)

# define constants
h = 6.626e-34
c = 2.9979e8
k = 1.3806e-23
```

```python
# Planck's blackbody radiation law
def bb(v, T): 
  bb = 8*np.pi*h*v**3 / (c**3 * (np.exp(h*v/(k*T)) - 1))
  return bb

# Rayleigh-Jeans' blackbody radiation law
def rj(v, T):
  rj = 8*np.pi*k*T*v**2 / c**3
  return rj
```

```python
# set up plot, make all fonts and lines bigger
fig = plt.figure(figsize=[20,12])
plt.rcParams.update({'font.size': 14, 'lines.linewidth':3})

# plot on log-log scale
plt.loglog(v,bb(v,300),label='300 K')
plt.loglog(v,bb(v,2000), label='2000 K')
plt.loglog(v,bb(v, 7000), label='7000 K')
plt.loglog(v,rj(v, 2000),'--k', label='Rayleigh-Jeans, 2000 K')

# set limits, add axis labels and legend
plt.ylim(1e-21,1e-14)
plt.xlim(1e11,1e16)
plt.ylabel('Energy Density')
plt.xlabel('Frequency (Hz)')
plt.legend()
```
