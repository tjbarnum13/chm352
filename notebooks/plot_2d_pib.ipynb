```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
def run(nx,ny,a,b):
  
  # set up grid and compute eigenfunction
  xs = np.linspace(0,a,1001)
  ys = np.linspace(0,b,1001)
  (x,y)=np.meshgrid(xs,ys)
  psi = np.sin(nx*np.pi*x/a)*np.sin(ny*np.pi*y/b)

  # calculate energy
  Ex=1/8*(nx**2/a**2)
  Ey=1/8*(ny**2/b**2)
  E=Ex+Ey

  # plot eigenstates
  plt.figure(figsize=(10, 6))
  plt.contourf(x,y,psi,cmap='seismic')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(f'E=E$_x$+E$_y$={Ex:.2f}+{Ey:.2f}={E:.2f} au')

  plt.tight_layout()

  plt.show()
```

```python
cu = False # turn off continous update for smoother operation

# set up all the sliders to change variables
nx=widgets.IntSlider(min=1, max=20, step=1, value=3, description='nx:',continuous_update=cu)
ny=widgets.IntSlider(min=1, max=20, step=1, value=3, description='ny:',continuous_update=cu)
a=widgets.FloatSlider(min=0.1, max=10, step=0.1, value=1.0, description='a:',continuous_update=cu)
b=widgets.FloatSlider(min=0.1, max=10, step=0.1, value=1.0, description='b:',continuous_update=cu)

# define the output and arrange sliders and plot
out = widgets.interactive_output(run,{'nx':nx,'ny':ny,'a':a,'b':b})
controls = widgets.VBox([nx, ny, a, b])
display(widgets.HBox([controls, out]))
```
